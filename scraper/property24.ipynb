```python
from bs4 import BeautifulSoup
import requests
import json

crib = []

url = 'https://www.property24.co.ke/property-to-rent-in-nairobi-c1890?ToPrice=50000'
data = requests.get(url).text

soup = BeautifulSoup(data, 'lxml')
houses = soup.find_all('div', class_='pull-left sc_listingTileContent')

for house in houses:
#     rent = house.find('div', class_='sc_listingTilePrice primaryColor').span.text.replace(' ', '')
#     location = house.find('div', class_='sc_listingTileAddress primaryColor').text.replace(' ', '')
#     description = house.find('div', class_='sc_listingTileTeaser').text
    try:
        location = house.find('div', {'class':'sc_listingTileAddress primaryColor'}).text.strip()
        rent = house.find('div', {'class':'sc_listingTilePrice primaryColor'}).span.text.strip()
#         description = house.find('div', {'class':'sc_listingTileTeaser'}).text

        print(location)
        print(rent)
#         print(description)
        print()

        dict = {}
        dict['location'] = location
        dict['rent'] = rent
#         dict['description'] = description

        crib.append(dict)
    except Exception as ex:
        print('Error:', ex)
        
f = open('data.json', 'w')
f.write(json.dumps(crib, indent=2))
f.close()
    
```
